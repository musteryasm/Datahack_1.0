```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/preprocessed_data(final).csv')
df.head()
```

```python
orders_df = df[['order_id','item_name']]
orders_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
orders_df['encoded_order_id'] = le.fit_transform(orders_df['order_id'])
```

```python
orders_df
```

```python
unique_item_name = orders_df['item_name'].unique()
```

```python
unique_item_name = pd.DataFrame(unique_item_name)
```

```python
unique_item_name.to_csv('/content/drive/MyDrive/Colab Notebooks/unique_item_name.csv')
```

```python
grouped_df = orders_df.groupby('encoded_order_id')['item_name'].apply(list)
```
