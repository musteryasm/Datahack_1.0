```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('/content/drive/MyDrive/DataHack /jetson-sample-data.csv')
data.head(5)
```

```python
data.dropna()
```

drop rows where value is negative

```python
df = data.drop(data[data['price'] < 0.0].index,inplace = True)
```

drop rows where item_name is ALL IN ON SAVORY , CCard Conveience , Staff Discount , cake cutting charge , 

```python
df = data = data.drop(data[data['item_name'].str.contains('ALL IN ON SAVORY|CCard Convenience|Staff Discount|Cake Cutti|Upcharge|Cc Charge|Cc fee|For Cc Tip|Pep Fee|Pep Charge|Pep', na=False)].index)
```

export csv file 

```python
df.to_csv('/content/drive/MyDrive/Colab Notebooks/preprocessed_data(final).csv')
```

```python
group_data_df = df[['order_id','client_id','date']]
group_data_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
group_data_df['encoded_order_id'] = le.fit_transform(group_data_df['order_id'])
group_data_df['encoded_client_id'] = le.fit_transform(group_data_df['client_id'])
```

```python
group_data_df
```

```python
new_df = pd.DataFrame()
new_df['client_id'] = group_data_df['encoded_client_id'].unique() 
new_df['grouped_ordered_id'] = group_data_df.groupby('encoded_client_id')['encoded_order_id'].unique().apply(list)
new_df['grouped_date'] = group_data_df.groupby('encoded_client_id')['date'].unique().apply(list)
new_df
```
